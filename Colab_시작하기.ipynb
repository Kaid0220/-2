```
def conditional_probability(prob_a_and_b, prob_b):
    return prob_a_and_b / prob_b

prob_a = 0.6
prob_b = 0.4
prob_a_and_b = 0.3

prob_a_given_b = conditional_probability(prob_a_and_b, prob_b)
print("우유를 산 경우에 빵을 살 확률:", prob_a_given_b)
```

```
import matplotlib.pyplot as plt
import numpy as np

colors = ['#348ABD', '#A60628']
prior = [1/25, 24/25]
posterior = [0.158, 0.158]

plt.bar([0, 0.7], prior, alpha=0.7, width=0.25, color=colors[0], label='사전확률', lw=3)
plt.bar([0+0.25, 0.7+0.25], posterior, alpha=0.7, width=0.25, color=colors[1], label='사후확률', lw=3)

plt.xticks([0.12, 0.82], ['가수', '직장인'])
plt.title("철수의 직업에 대한 사전확률과 사후확률")
plt.ylabel("확률")
plt.legend(loc='upper left')
plt.show()






```
