```
import matplotlib.pyplot as plt

joint_pmf_data = {
    (1, 1): 0.1,
    (1, 2): 0.2,
    (1, 3): 0.2,
    (2, 1): 0.3,
    (2, 2): 0.05,
    (2, 3): 0.15
}

f_X = {}
f_Y = {}

for x, y in joint_pmf_data.keys():
    f_X[x] = f_X.get(x, 0) + joint_pmf_data[(x, y)]

for x, y in joint_pmf_data.keys():
    f_Y[y] = f_Y.get(y, 0) + joint_pmf_data[(x, y)]

plt.figure(figsize=(12, 5))
y_min, y_max = 0, 1
x_min, x_max = 0.5, 3.5

plt.bar(f_X.keys(), f_X.values(), width=0.3, color='skyblue', label='f_X(x)')
plt.bar([y + 0.3 for y in f_Y.keys()], f_Y.values(), width=0.3, color='orange', label='f_Y(y)')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xlabel('x, y')
plt.ylabel('Probability')
plt.title('결합분포로부터 계산한 X와 Y의 주변확률질량함수')
plt.legend()
plt.show()
```

```
import numpy as np

korean_scores = [75, 65, 50, 70, 80]
math_scores = [65, 70, 70, 55, 60]

data = np.array([korean_scores, math_scores])
covariance = np.cov(data, bias=True)[0, 1]

print("Covariance:", covariance)
```
