```
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5, 6, 7])

y = np.array([78.4, 78.6, 78.5, 78.9, 78.2, 79.4, 78.6])

plt.plot(x, y, color="r")

plt.title("7일간 체중 변화")
plt.xlabel("날짜")
plt.ylabel("체중 (kg)")
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5, 6, 7])

y = np.array([78.4, 78.6, 78.5, 78.9, 78.2, 79.4, 78.6])

plt.scatter(x, y, color="r")

plt.title("7일간 체중 변화 (산점도)")
plt.xlabel("날짜")
plt.ylabel("체중 (kg)")

plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5, 6, 7])

y = np.array([78.4, 78.6, 78.5, 78.9, 78.2, 79.4, 78.6])

plt.plot(x, y, color='r', marker='o')

plt.grid()

plt.title("7일간 체중 변화 (그리드 포함)")
plt.xlabel("날짜")
plt.ylabel("체중 (kg)")

plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 300)

y1 = x ** 2
y2 = (x - 2) ** 2

plt.plot(x, y1, color="r")
plt.plot(x, y2, color="k", linestyle="--")

plt.grid(True)
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['axes.unicode_minus'] = False

x = np.linspace(-5, 5)
y1 = 2 * x
y2 = 5 * x

plt.plot(x, y1, label="y1 값", color='r')
plt.plot(x, y2, label="y2 값", linestyle="dashed", color='k')

plt.xlabel("X-축", size=14)
plt.ylabel("Y-축", size=14)
plt.title("그래프 제목")
plt.grid()
plt.legend()

plt.show()
```
